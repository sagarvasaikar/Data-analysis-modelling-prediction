# Import Data

```python
import numpy as np
import pandas as pd

url="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data" #Load Data
df= pd.read_csv(url,header= None)  
```

```python
df.head() 
```

```python
#Giving column names
headers= ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns=headers

#SAVES TO CSV
df.to_csv("automobile.csv",index=False) 
```

# Data Cleaning

```python
df.isnull().any().any()  #check null values  
```

```python
df.isnull().sum()
```

```python
#Convert ? into Nan values in table
df.replace("?",np.nan, inplace=True)
df.head()
```

```python
missing_data = df.isnull()
missing_data.head(5)  #shows missing values
```

```python
#Calculating missing values in each column

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
#Here we are filling up colums with the mean value of all emelents in that column

avg=df['normalized-losses'].astype(float).mean()
print(avg)

avg1=df['bore'].astype(float).mean()
print(avg1)

avg2=df['stroke'].astype(float).mean()
print(avg2)

avg3=df['horsepower'].astype(float).mean()
print(avg3)

avg4=df['peak-rpm'].astype(float).mean()
print(avg4)
```

```python
df['normalized-losses'].replace(np.nan,avg,inplace=True)
df['bore'].replace(np.nan,avg1,inplace=True)
df['stroke'].replace(np.nan,avg1,inplace=True)
df['horsepower'].replace(np.nan,avg1,inplace=True)
df['peak-rpm'].replace(np.nan,avg1,inplace=True)
```

```python
#Here we will count no of entries in 2door cars and 4door cars
df['num-of-doors'].value_counts()
```

```python
#Here we will see the max value in the column
df['num-of-doors'].value_counts().idxmax()
```

```python
#Here we replace it will the value which is repeated frequently 
df['num-of-doors'].replace(np.nan,"four",inplace=True)
```

```python
#let's drop all rows that do not have price data:
df.dropna(subset=["price"],axis=0,inplace=True)
# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

# Data Formatting

```python
#we will check the datatypes of all columns

df.dtypes
```

```python
#We will now convert the datatypes 

df['price']=df['price'].astype("float")
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
#Lets check the datatypes
df.dtypes
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

# Data Normalisation 

```python
#We will normalise the length,width,height column by using simple feature scaling method

df['length']=df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

# Binning

```python
bins=np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#We will convert categorical values to numerical
pd.get_dummies(df['fuel-type'])
df['fuel-type'].rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
```
